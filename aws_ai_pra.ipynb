**biases can affect the fairness, accuracy, and trustworthiness of the modelâ€™s predictions**


**Algorithmic Bias**
Algorithmic bias occurs when the machine learning (ML) algorithm itself introduces or amplifies biases, leading to unfair or skewed outcomes. This bias can stem from the design or operation of the algorithm, regardless of the data used.

 **Sample Bias (Sampling Bias)**
 Sample bias, or sampling bias, occurs when the data collected or selected for training the model does not adequately represent the population or use case. This results in the model making predictions that are biased toward certain groups or categories


**Measurement Bias**
Measurement bias occurs when there are errors or inconsistencies in how features or outcomes are measured. This can lead to the model learning from incorrect or skewed data, which results in biased predictions.

**Recency Bias**:-Recency bias occurs when a model or system disproportionately gives more weight or importance to recent events or data points, while undervaluing older or historical data
